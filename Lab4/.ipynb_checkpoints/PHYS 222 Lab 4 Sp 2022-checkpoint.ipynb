```python
### SOLUTION: REMVOE FROM STUDENT VERSION!

def load_waveform(filename, showfig=True) :
    data = np.genfromtxt(filename, skip_header=8, delimiter=",")
    ncol = data.shape[1]
    #print(ncol)
    t = data[:,1]
    ch1 = data[:,2]
    
    if showfig == True :  # only generate a new figure if the newfig flag is true
        plt.plot(t, ch1, label = "Channel 1")
    
        if ncol == 4 : # this means both Ch1 and Ch2 were recorded
            ch2 = data[:,3]
            plt.plot(t, ch2, label = "Channel 2")

        plt.xlabel("Time (s)")
        plt.ylabel("Voltage (V)")
        plt.legend()
    if ncol == 4 : 
        return [t, ch1, ch2]
    elif ncol == 3 :
        return [t, ch1]
```

# Task 3: Frequency dependent impedance of a capacitor
This task extends the analysis of the Task 2 to multiple frequencies. The goal is to make measurements that verify the complex-valued impedance formula for a capacitor.

$$ X_C = \frac{1}{\omega C} $$

Because $X_C$, relates current to voltage similar to Ohm's law, we have 

$$V_C = -j X_C I_C = -j \frac{1}{\omega C} I_C $$

#### We will be able to test 3 main features of this formula:
* The imaginary $i$ means the sinusoidal voltage across capacitor is 90 degrees behind of the sinusoidal current. There is a 90 degree phase shift between them.
* The magnitude of impedance decreases like $1/\omega$
* The magnitude is inversely proportional to capacitance $C$

#### Suggested steps

1. Calculate the frequency (in Hz) where the capacitive reactance $X_C$ has the same magnitude value as the resistance $R$. We will call this $f_{RC}$.  
1. Select drive frequencies that span well below $f_{RC}$ to a bit above. I picked values approximately $0.01 f_{RC}$ up to $10f_{RC}$. Use a log-scale for spacing frequencies, such as 100, 300, 1000, 3000, etc.
1. Measure the time-shift between the two waves and use the cursors to measure the $\Delta t$
1. Measure the amplitude of the Ch1 and Ch2 sine waves. $V_{pp}$ - peak-to-peak amplitude is a good choice. It can be read using the "Measurement tools
1. Make a table of the measurements

```python
#### SOLUTION
data = np.genfromtxt("Lab4_XC_data_Scopy_1kOhm.csv", delimiter=",", skip_header=3)
f = data[:,0]
tshift = data[:,1]
VRpp = data[:,2]
VCpp = data[:,3]

phase_deg = 360 * f * (tshift)

R = 1e3
Ipp = VRpp / R

XC = VCpp / Ipp

print(phase_deg)

# plot the phase shift between VC and current current (VR)
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.semilogx(f, phase_deg,"o", label="Measured")
plt.grid()
plt.ylim(-180,180)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase shift (degrees)")
plt.title("Example: Phase shift is close to -90 degrees for all frequencies")

ftheory = np.linspace(100, 3e5, 1000)
phase_deg_theory = -90 * np.ones(len(ftheory))
plt.semilogx(ftheory,phase_deg_theory, "--", label="Theory")
plt.legend()

# plot the magnitude of X_C
plt.subplot(1,2,2)
plt.loglog(f,XC, "o", label="Measured")
plt.xlabel("Frequency (Hz)")
plt.ylabel("$X_C$ (Ohms)")


C = 9.9e-9  # 10 nF capacitance nominal
ftheory = np.linspace(100, 3e5, 1000)
XCtheory = 1/(2*np.pi*ftheory * C)
plt.loglog(ftheory, XCtheory, label="Theory")
plt.title("Example: Impedance decreases like 1/$\omega C$")
plt.grid()
plt.legend()
plt.savefig("example_XC_plots.png")

print("Frequency ", f)
print("XC data   ", XC)
print("XC theory ", 1./(2*np.pi*f * C))
```

# Task 4: Frequency dependence of the inductor
![RL circuit for measuring impedance of inductor](Fig_RL_Task4.png)

This is similar to Task 3, but for the inductor. The goal is to make measurements that demonstrate the complex-valued impedance formula for an inductor.

$$ X_L = \omega L $$

Because $X_L$, relates current to voltage similar to Ohm's law, we have 

$$V_L = j X_L I_L = j \omega L I_L $$

#### We will be able to test 3 main features of this formula:
* The imaginary $-i$ means the sinusoidal voltage across inductor is 90 degrees ahead of the sinusoidal current. There is a 90 degree phase shift between them.
* The magnitude of impedance increases with $\omega$
* The magnitude is proportional to inductance $L$
* The formula breaks down for small $\omega$ because the inductor is not ideal

#### Suggested steps
1. Build the circuit. 
1. At what frequency (in Hz) should $X_L$ and $R$ have the save value?

1. Use the cursors to measure the $\delta t$ between the Voltage across the inductor and resistor. Note which is leading.
1. Measure the peak-to-peak amplitude $V_{pp}$ is a good choice.

```python



#### SOLUTION
data = np.genfromtxt("Lab4_XL_data.csv", delimiter=",", skip_header=1)
f = data[:,0]
VRpp = data[:,1]
tshift_R = data[:,2]
VLpp = data[:,3]
tshift_L = data[:,4]


#### ASHLEYS DATA, columns ordered differently
# data = np.genfromtxt("Lab_4_2_ashley.csv", delimiter=",", skip_header=1)
# f = data[:,0]
# VLpp = data[:,2]
# tshift_L = data[:,1]*1e-3  # convert ms to s
# VRpp = data[:,4]
# tshift_R = data[:,3]*1e-3  # convert ms to s



# phase shift between current (prop. to VR) and inductor voltage VLpp
phase_deg = 360 * f * (-tshift_L + tshift_R)

R = 100
Ipp = VRpp / R

XL = VLpp / Ipp

print(phase_deg)

# plot the phase shift between VC and current current (VR)
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.semilogx(f, phase_deg,"o", label="Measured")
plt.grid()
plt.ylim(-180,180)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase shift (degrees)")
plt.title("Example: Phase shift between inductor voltage and current")

ftheory = np.linspace(100, 100000, 1000)
phase_deg_theory = 90 * np.ones(len(ftheory))
plt.semilogx(ftheory,phase_deg_theory, "--", label="Ideal inductor (theory)")
plt.legend()

# plot the magnitude of X_L
plt.subplot(1,2,2)
plt.loglog(f,XL, "o", label="Measured")
plt.xlabel("Frequency (Hz)")
plt.ylabel("$X_L$ (Ohms)")

L = 10e-3  # 10 mH inductance nominal
ftheory = np.linspace(100, 100000, 1000)
XLtheory = 2*np.pi*ftheory * L
plt.loglog(ftheory, XLtheory, '--', label="Ideal inductor (theory)")

RLtheory = 40 * np.ones(len(ftheory))  # 40 ohm resistance of inductor
plt.loglog(ftheory, RLtheory, '--', label="Internal resistance of inductor")
plt.title("Example: Impedance increases like $\omega L$")
plt.grid()
plt.legend()
plt.savefig("example_XL_plots.png")
```
