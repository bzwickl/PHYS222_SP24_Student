# Task 1: Copy your `load_waveform` function from Lab 3

**Insert cell or cells below to record your answers**

```python
### SOLUTION: REMVOE FROM STUDENT VERSION!

def load_waveform(filename, showfig=True) :
    data = np.genfromtxt(filename, skip_header=8, delimiter=",")
    ncol = data.shape[1]
    #print(ncol)
    t = data[:,1]
    ch1 = data[:,2]
    if ncol == 4 : # this means both Ch1 and Ch2 were recorded
            ch2 = data[:,3]
    
    if showfig == True :  # only generate a new figure if the newfig flag is true
        plt.plot(t, ch1, label = "Channel 1")
    
        if ncol == 4 : # this means both Ch1 and Ch2 were recorded
            plt.plot(t, ch2, label = "Channel 2")

        plt.xlabel("Time (s)")
        plt.ylabel("Voltage (V)")
        plt.legend()
    if ncol == 4 : 
        return [t, ch1, ch2]
    elif ncol == 3 :
        return [t, ch1]
```

# Task 2: Build the 10x inverting amplifier op-amp circuit

This lab uses a general purpose op-amp, the Texas Instruments UA741. You can download the full [data sheet and specifications](https://www.ti.com/lit/ds/symlink/ua741.pdf) from TI. 

![10x inverting amplified schematic](Fig_Lab7_Task2.png)

**This circuit uses 5 out of the 8 pins:**
* In+, labeled +  on the schematic, is the "non-inverting input"
* In-, labeled - on the schematic, is the "inverting input"
* Out, is the output
* VCC+ is the positive supply voltage. In our case, the M2K V1 (red wire) is set to +5V
* VCC- is the negative supply voltage. In our case, the M2K V2 (white wire) is set to -5V

# Task 4: Slew rate

Real world op-amps can't change their voltage arbitrarily fast. The maximum rate (often specified in volts per microsecond) is known as the slew rate of the op-amp. According to the UA741 data sheet, the slew rate is around 0.5 V/$\mu$s = 500,000 V/s. However the data sheet test circuit was a gain = 1 amplifier, so your measured slew rate could be different by a factor of 2 or more.

1. Change the input to a **square wave** (It is the waveform with fastest rise/fall time). 
1. Set amplitude to something larger than 100 mVpp, but don't saturate the output. 
1. If you set the frequency higher, between 20 kHz - 100 kHz, you should start to see behavior where the amplifier clearly doesn't reproduce the square wave. It is most noticable at higher frequencies. 
1. Import and plot a data set that shows the slew rate.
1. In Scopy, use cursors or any other method to measure the slope. The "track" cursors can give both a $\Delta x$ and $\Delta y$ with one measurement.

**Insert cell or cells below to record your answers**

```python
# Falling (measured with cursors)
dt = 4.072e-6 # seconds
dv = -1.401 # volts

slew_falling = dv / dt
print("Slew (falling) (V/s) = ", slew_falling)

#rising
dt = 3.967e-6 # seconds
dv = 1.377 # volts

slew_rising = dv / dt
print("Slew (rising) (V/s) = ", slew_rising)

load_waveform("Sol_new_Lab7_Task4.csv")
plt.xlim(0,4e-5)
plt.title("100 kHz, 300 mVpp square wave")
```

The spec'd slew rate from the data sheet was 0.5 V / uS at unity gain.  

The measured slew rate for a 10x inverting was around 0.35 V / uS.

# Task 5: Gain vs frequency (for 10x inverting)

The inverting amplifier works well at lower frequencies, but not so well at high frequenices. This task will characterize the frequency response of the inverting amplifier.

1. Set the waveform to a sine wave with 1 kHz and 300 mVpp amplitude (as large as possible without saturating the amplifier)
1. Measure the input and ouput amplitudes at 1, 2, 5, 10, 20, 50, 100, 200, 500 kHz. 
    * Making a CSV data table is recommended.
1. Calcualte the gain (dB) using your data. Make a gain vs frequency plot. 
1. Based on the graph, determine the 3 dB frequency, which is where the gain decreases below the ideal (low-frequency) gain by 3 dB. This corresponds to 1/2 power or $1/\sqrt{2} \approx 0.707$ times amplitude of the low frequency gain.
1. Determine the gain-bandwidth product, which is the product of the 3 dB frequency and the gain at that same frequency (as a multiplier, not as dB). 
    * The data sheet shows a gain-bandwidth product of about 1 MHz. How does your data compare?
1. What is the phase shift between input and output for a frequency well below the 3 dB point (e.g., 1 kHz) and a frequency well above (e.g., 200 kHz)?

**Insert cell or cells below to record your answers**

```python
#### SOLUTION: 
data = np.genfromtxt("Sol_new_Lab7_Task5.csv", delimiter = ",", skip_header=8)

# extract the data from each column
fdata = data[:,1]
dBdata = data[:,2]
phasedata = data[:,3]


# Make the gain data plot
plt.semilogx(fdata, dBdata, "o", label = "Experiment")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.grid()
plt.title("10x inverting amplifier")

# Make the gain data plot
plt.figure()
plt.plot(fdata, dBdata, "o", label = "Experiment")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.grid()
plt.title("10x inverting amplifier")
plt.xlim(4e4,6e4)
plt.ylim(15,20)


# Make the phase plot
plt.figure()
plt.semilogx(fdata, phasedata, "o", label = "Experiment")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (deg)")
plt.grid()
plt.title("10x inverting amplifier")


gain17dB = np.power(10,17 / 20)
print("Gain (as number) at 17 dB = ", gain17dB)
f3db = 48e3  # 3 dB frequnency 40 kHz (see graph)

gain_bw = gain17dB * f3db 
print("Gain bandwidth product  = ", gain_bw)

## Phase shift at 200 kHz
#read off graph
phase = 90
print("phase shift at 200 kHz = ", phase)
```

The 3db point in the gain curve appears to be around 40 kHz (crosses 17 dB line).

The phase shift at high frequencies is 90 degrees, whereas at low frequencies it is 180 degrees. Similar to a capacitor, there is a 90 degree phase shift between low and high frequencies.

1. How does the gain-bandwidth product compare for the 10x and 21x inverting amplifiers?
 

**Insert cell or cells below to record your answers**

```python
#### SOLUTION: 
data = np.genfromtxt("Sol_new_Lab7_Task6.csv", delimiter = ",", skip_header=8)

# extract the data from each column
fdata = data[:,1]
dBdata = data[:,2]
phasedata = data[:,3]


# Make the gain data plot
plt.semilogx(fdata, dBdata, "o", label = "Experiment")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.grid()
plt.title("21x inverting amplifier")

# Make the gain data plot
plt.figure()
plt.plot(fdata, dBdata, "o", label = "Experiment")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB)")
plt.grid()
plt.title("21x inverting amplifier")
plt.xlim(1e3,6e4)
plt.ylim(21,27)


# Make the phase plot
plt.figure()
plt.semilogx(fdata, phasedata, "o", label = "Experiment")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (deg)")
plt.grid()
plt.title("21x inverting amplifier")


gain17dB = np.power(10,26.5 / 20)
print("Gain (as number) at 17 dB = ", gain17dB)
f3db = 26e3  # 3 dB frequnency 40 kHz (see graph)

gain_bw = gain17dB * f3db 
print("Gain bandwidth product  = ", gain_bw)

## Phase shift at 200 kHz
#read off graph
phase = 90
print("phase shift at 200 kHz = ", phase)
```

The gain at low frequencies was around 25.3 dB, the 3 dB point is around 23 kHz (a bit hard to read off the log scale).
However it shows that when the gain went up by a factor of 2, the bandwidth decreased by a factor of 2.

* In task 5, Gain-BW product was 280 kHz
