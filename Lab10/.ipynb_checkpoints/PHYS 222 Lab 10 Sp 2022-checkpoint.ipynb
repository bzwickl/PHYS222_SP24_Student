### Tips

* You will use the LED light source from Lab 9 Task 6 to illuminate your finger.
* Place your finger on the photodiode. A smaller finger might work better...but any could work.
* If your photodiode comes loose from the breadboard. Use a small piece of transparent tape to hold it down. 
* Touch the LED to the top of your finger, your finger will probably glow red. 
* Adjust the LED position so it doesn't saturate the transimpedance amplifier stage, but still provides 2-3 V of signal.
* Keep your finger and LED really **really steady.** Smooth, gradual changes in alignment work best. Abrupt changes in alignment will cause the amount of detected light to change quickly and produce fluctuations in the signal that are not caused by a heart beat.
